## Введение

### Описание проекта:

Перед интернет-магазином «В один клик» стоит амбициозная цель: удержать существующих клиентов и повысить их вовлечённость. Чтобы достичь этого, необходимо персонализировать взаимодействие с каждым покупателем, создавая предложения, которые будут максимально соответствовать их интересам и потребностям.

Проект направлен на автоматизацию и оптимизацию процесса персонализации с использованием анализа данных. Задачи включают изучение покупательской активности, построение модели прогнозирования предпочтений клиентов и разработку рекомендаций для увеличения их активности. Все действия будут подкреплены аналитическими выводами, чтобы минимизировать риски и обеспечить успех инициативы.

### Описание данных

Данные для работы представлены в нескольких таблицах.

`market_file.csv` 

Содержит информацию о поведении покупателя на сайте, его активности и взаимодействиях:  
- **id** — уникальный идентификатор покупателя.  
- **Покупательская активность** — целевой признак: «снизилась» или «прежний уровень».  
- **Тип сервиса** — уровень сервиса: «премиум» или «стандарт».  
- **Разрешить сообщать** — согласие на получение рекламных предложений.  
- **Маркет_актив_6_мес** — среднемесячное число маркетинговых коммуникаций за последние 6 месяцев (e-mail, звонки, реклама и т.д.).  
- **Маркет_актив_тек_мес** — количество коммуникаций в текущем месяце.  
- **Длительность** — количество дней с момента регистрации покупателя.  
- **Акционные_покупки** — среднемесячная доля покупок по акциям за последние 6 месяцев.  
- **Популярная_категория** — самая популярная категория товаров покупателя за последние 6 месяцев.  
- **Средний_просмотр_категорий_за_визит** — среднее число просмотренных категорий за визит за последний месяц.  
- **Неоплаченные_продукты_штук_квартал** — количество товаров в корзине, оставшихся неоплаченными за последние 3 месяца.  
- **Ошибка_сервиса** — количество сбоев/ошибок, затронувших покупателя.  
- **Страниц_за_визит** — среднее число страниц, просмотренных за визит за последние 3 месяца.

`market_money.csv` 
Содержит информацию о выручке, полученной с конкретного покупателя:

- **Выручка** — сумма выручки за указанный период.

Данные о времени, проведённом покупателем на сайте за указанный период:

- **Период** — период времени («текущий_месяц», «предыдущий_месяц»).  
- **минут** — общее время на сайте, в минутах.  

---

`money.csv`
Среднемесячная прибыль, которую магазин получает от покупателя за последние 3 месяца:

- **Прибыль** — значение среднемесячной прибыли.

````python
# Функция для вывода первых строк и общей информации о датафрейме
def chek_df(df, name):
    """
    Выводит первые строки и информацию о переданном датафрейме.
    
    Параметры:
    - df: датафрейм, который нужно изучить
    - name: имя датафрейма (строка), чтобы его можно было идентифицировать в выводе
    """
    print(f"--- {name} ---")
    display("Первые строки:", df.head(), "\n")
    print("Информация:")
    print(df.info(), "\n")
    print("Наличие пропусков:")
    print(df.isna().sum())
    print("Дубликатов в данных:", df.duplicated().sum())
    print("="*40)  # Разделитель для визуального удобства
    
#=========================================================#

# Новая функция с учетом дискретных значений

def histnbox_plot(df):
    """
    Функция для построения графиков распределения количественных данных: гистограмм, ящиков с усами и countplot.
    
    df: DataFrame – исходный датасет, для которого необходимо построить графики.
    """
    # Проверка наличия количественных данных в датасете
    numeric_columns = df.select_dtypes(exclude=["object"]).columns
    if numeric_columns.empty:
        print("В датасете отсутствуют количественные данные для визуализации!")
        return
    
    # Устанавливаем стиль графиков
    sns.set_style('whitegrid')
    
    # Общий заголовок графиков
    title = f'Графики распределения для количественных данных датасета: {getattr(df, "name", "noname")}'
    
    # Настройка размеров фигуры
    num_plots = len(numeric_columns)  # Количество количественных признаков
    fig, axes = plt.subplots(num_plots, 2, figsize=(16, num_plots * 5))
    
    # Приведение массива подграфиков к двумерной форме (даже если только один признак)
    if num_plots == 1:
        axes = np.expand_dims(axes, axis=0)
    
    # Настройка параметров шрифтов
    plt.rcParams.update({'font.size': 10}) 
    
    # Построение графиков
    for i, col in enumerate(numeric_columns):
        # Определить, является ли признак дискретным или непрерывным
        unique_values = df[col].nunique()
        if unique_values / len(df) < 0.1:  # Если доля уникальных значений мала, считаем дискретным
            data_type = "discrete"
        else:
            data_type = "continuous"
        
        if data_type == "discrete":
            # Построение countplot для дискретных данных
            sns.countplot(x=df[col], ax=axes[i, 0], palette="pastel")
            axes[i, 0].set_title(f'Countplot: {col} (тип: {df.dtypes[col]})')
            axes[i, 0].set_ylabel('Частота')
            axes[i, 0].set_xlabel(f'{col}')
            
            # Настройка подписи на оси X, чтобы избежать наложения
            axes[i, 0].tick_params(axis='x', rotation=45)  # Угол наклона меток по X
            axes[i, 0].set_xticklabels(axes[i, 0].get_xticklabels(), ha='right')  # Выравнивание меток по правому краю
        else:
            # Гистограмма для непрерывных данных
            sns.histplot(data=df[col], kde=True, bins=50, ax=axes[i, 0])
            axes[i, 0].set_title(f'Гистограмма: {col} (тип: {df.dtypes[col]})')
            axes[i, 0].set_xlabel(f'{col}')
            axes[i, 0].set_ylabel('Количество')
        
        # Boxplot-только для непрерывных данных
        if data_type == "continuous":
            sns.boxplot(y=df[col], ax=axes[i, 1],
                        showmeans=True,
                        meanprops={"marker": "s", "markerfacecolor": "white", "markeredgecolor": "blue", "markersize": "5"})
            axes[i, 1].set_title(f'Boxplot: {col}')
            axes[i, 1].set_xlabel(f'{col}')
        else:
            axes[i, 1].remove()  # Убираем лишние графики для дискретных данных
    
    # Общий заголовок
    plt.suptitle(title, fontsize=20, y=1.01)
    plt.tight_layout()  # Автоматическое упорядочивание графиков
    plt.show()

#=========================================================#

def show_bar_for_categorical(df):
    """
    Строит столбчатые диаграммы для всех категориальных признаков DataFrame,
    располагая их на сетке графиков.

    Parameters:
    df: DataFrame, содержащий данные.
    """
    # Выбираем только категориальные столбцы (тип object или category)
    categorical_columns = df.select_dtypes(include=['object', 'category']).columns
    num_plots = len(categorical_columns)

    # Если категориальных столбцов нет, выводим предупреждение
    if num_plots == 0:
        print("В таблице отсутствуют категориальные признаки.")
        return

    # Определяем размер сетки (rows x cols) для подграфиков
    cols = 2  # Количество столбцов в сетке (можно менять)
    rows = math.ceil(num_plots / cols)

    # Создаем фигуру и множество осей
    # Убираем constrained_layout=True, чтобы не было конфликта с tight_layout
    fig, axes = plt.subplots(rows, cols, figsize=(12, rows * 5))

    # Преобразуем оси в одномерный массив (если нужно), чтобы простой итерацией заполнить графики
    axes = axes.flatten()

    # Строим графики по каждому столбцу
    for i, name in enumerate(categorical_columns):
        pivot = df[name].value_counts().sort_values(ascending=True)

        pivot.plot.bar(
            ax=axes[i],
            xlabel=name,
            ylabel='Количество',
            title=f'Распределение для {name}',
            rot=45  # Поворачиваем подписи на оси X
        )
        # Настраиваем шрифт заголовка графика
        axes[i].title.set_fontsize(12)

    # Удаляем лишние пустые графики в сетке, если столбцов меньше, чем ячеек в сетке
    for j in range(len(categorical_columns), len(axes)):
        fig.delaxes(axes[j])

    # Устанавливаем отступы между графиками с помощью tight_layout
    plt.tight_layout()

    # Показываем итоговую фигуру с сеткой
    plt.show()

#=========================================================#

def dist_plot_for(df):
    """
    Функция для построения графиков распределения числовых признаков в разрезе категориальной переменной.
    
    df: DataFrame – датасет, содержащий числовые признаки и категориальный признак "покупательская_активность".
    """
    # Установка стиля для графиков
    sns.set_style('whitegrid')
    
    # Заголовок для всего набора графиков
    title = 'Распределение числовых признаков в разрезе пользовательской активности'
    
    # Выбираем только числовые колонки
    numeric_columns = df.select_dtypes(exclude=["object"]).columns
    if numeric_columns.empty:
        print("В датасете нет числовых данных для анализа!")
        return

    # Вычисляем количество строк для подграфиков
    num_plots = len(numeric_columns)
    rows = int(np.ceil(num_plots / 2))  # Две гистограммы на строку

    # Создание сетки графиков
    fig, axes = plt.subplots(rows, 2, figsize=(16, rows * 6))

    # Обработка случая, если только один ряд (чтобы axes всегда был двумерным)
    if rows == 1:
        axes = np.expand_dims(axes, axis=0)

    # Настройка размера шрифта для подписей
    plt.rcParams.update({'font.size': 10})

    # Построение графиков
    for i, col in enumerate(numeric_columns):
        # Определяем положение текущего графика на сетке
        row, col_idx = divmod(i, 2)

        # График распределения числового признака в разрезе "покупательской активности"
        sns.histplot(data=df, x=col, hue="покупательская_активность", ax=axes[row, col_idx])
        
        # Настройка подписей и заголовков для каждого графика
        axes[row, col_idx].set_title(f'Распределение: {col}')
        axes[row, col_idx].set_xlabel(col)
        axes[row, col_idx].set_ylabel('Количество')

    # Убираем лишние пустые оси, если общее количество графиков нечётное
    if num_plots % 2 != 0:
        fig.delaxes(axes[-1, -1])  # Удаляем последнюю ось, если она осталась пустой

    # Общий заголовок для фигуры
    plt.suptitle(title, fontsize=20, y=1.01)
    plt.tight_layout()  # Делаем графики аккуратными
    plt.show()

#=========================================================#
````

````python
# Создаем пайплайн для категориальных признаков (ohe_columns)
ohe_pipe = Pipeline(
    [
        ("SimpleImputer_ohe", SimpleImputer(strategy="most_frequent")),
        (
            "ohe",
            OneHotEncoder(
                drop="first", handle_unknown="ignore"
            ),  # Удаляем первую категорию, поддерживаем неизвестные значения
        ),
    ]
)

# Обработчик ordinal признаков
ord_pipe = Pipeline(
    [
        ("SimpleImputer_ord", SimpleImputer(strategy="most_frequent")),
        (
            "ord",
            OrdinalEncoder(
                categories=[
                    [
                        "Товары для детей",
                        "Домашний текстиль",
                        "Косметика и аксесуары",
                        "Техника для красоты и здоровья",
                        "Кухонная посуда",
                        "Мелкая бытовая техника и электроника",
                    ],
                ],
                handle_unknown="use_encoded_value",
                unknown_value=-1,
            ),
        ),
    ]
)

# Общий препроцессор для всех колонок
data_preprocessor = ColumnTransformer(
    [
        ("ohe", ohe_pipe, ohe_columns),  # Для ohe_columns
        ("ord", ord_pipe, ord_columns),  # Для ord_columns
        ("num", StandardScaler(), num_columns),  # Для числовых признаков
    ],
    remainder="passthrough",
    verbose_feature_names_out=False,  # Убираем лишние префиксы в именах признаков
)

X_preprocessed = data_preprocessor.fit_transform(X_train)

# Выводим форму данных после препроцессинга
print("Форма данных после препроцессинга:", X_preprocessed.shape)
````

### Пайплайн обучения моделей

В данных наблюдается дисбаланс классов целевой переменной. Для его корректировки будет применено сэмплирование, так как использование взвешивания не подходит из-за наличия моделей, которые не поддерживают работу с весами. В данном случае будет использован метод оверсэмплинга — **SMOTENC**, который позволяет создавать новые примеры, учитывая как численные, так и категориальные признаки, без потери каких-либо данных.

**Формирование категориальной маски** необходимо для корректной работы **SMOTENC**, который умеет учитывать особенности категориальных признаков (например, преобразованных через OneHotEncoder). 

Поскольку после препроцессинга структура данных меняется (категориальные признаки расширяются из-за OHE, добавляются новые ord и числовые признаки), нам важно указать SMOTENC, какие столбцы в итоговой матрице являются **категориальными**. Без правильной маски он будет работать некорректно, не разделяя категории и численные данные. Мы создали маску, которая соответствует **обработанным данным** после `ColumnTransformer`, чтобы SMOTENC смог учитывать оригинальные категории при создании синтетических данных.

````python
# Формирование категориальной маски
categorical_mask_expanded = []

# Список имен новых признаков из OHE
ohe_feature_names = data_preprocessor.named_transformers_["ohe"]["ohe"].get_feature_names_out()

# Увеличиваем маску категориальных признаков для OneHotEncoder
categorical_mask_expanded.extend([True] * len(ohe_feature_names))

# Добавляем ординальные признаки
categorical_mask_expanded.extend([True] * len(ord_columns))

# Числовые признаки (маска False)
categorical_mask_expanded.extend([False] * len(num_columns))

# Проверяем длину маски
assert len(categorical_mask_expanded) == X_preprocessed.shape[1], "Длина categorical_mask не совпадает с количеством признаков"

print("Категориальная маска:", categorical_mask_expanded)
````

````python
# Создаем SMOTENC
smote = SMOTENC(
    categorical_features=categorical_mask_expanded, random_state=RANDOM_STATE
)

# Производим ресэмплинг
X_resampled, y_resampled = smote.fit_resample(X_preprocessed, y_train)

print("Форма после SMOTE:", X_resampled.shape)
print("Количество записей (y):", len(y_resampled))
````

SMOTENC успешно отработал и данные были увеличены для балансировки классов. Это отличный результат, и теперь следующим шагом можно перейти к созданию финального пайплайна.

````python
# Финальный пайплайн
pipe_final = ImbPipeline(
    steps=[
        # Шаг 1: Препроцессор (обработка категориальных, ординальных и числовых данных)
        ('preprocessor', data_preprocessor),
        
        # Шаг 2: SMOTE для балансировки классов
        ('smote', SMOTENC(
            categorical_features=categorical_mask_expanded,  # Категориальная маска после препроцессинга
            random_state=RANDOM_STATE
        )),
        
        # Шаг 3: Модель — Decision Tree Classifier
        ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))  # Дерево решений с фиксированным seed
    ]
)

print("Pipeline создан успешно!")
````

### Создание алгоритма поиска оптимальных гиперпараметров

````python
# Определяем гиперпараметры для каждой модели
param_grids = {
    # Гиперпараметры для KNeighborsClassifier
    'KNeighborsClassifier': {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': [3, 5, 7, 9],
        'models__weights': ['uniform', 'distance']
    },
    # Гиперпараметры для DecisionTreeClassifier
    'DecisionTreeClassifier': {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': [3, 5, 10, None],
        'models__criterion': ['gini', 'entropy']
    },
    # Гиперпараметры для LogisticRegression
    'LogisticRegression': {
        'models': [LogisticRegression(max_iter=1000)],
        'models__C': [0.01, 0.1, 1, 10],
        'models__penalty': ['l2'],
        'models__solver': ['liblinear', 'lbfgs']
    },
    # Гиперпараметры для SVC
    'SVC': {
        'models': [SVC(probability=True, random_state=RANDOM_STATE)],
        'models__C': [0.01, 0.1, 1, 10],
        'models__kernel': ['linear', 'rbf']
    }
}
````

### Обучение алгоритма и поиск лучшей модели

Для оценки качества работы модели мы будем использовать метрику **ROC-AUC** так как по целевому классу наблюдается дисбаланс плюс данная метрика учитывает работу модели при всех возможных значениях порогов и поэтому не зависит от одного конкретного. 

````python
# 3. Выполняем поиск гиперпараметров через GridSearchCV для каждой модели
best_models = {}
best_scores = {}

for model_name, params in param_grids.items():
    print(f"Оптимизация гиперпараметров для модели {model_name}...")

    # Настраиваем GridSearchCV
    search = GridSearchCV(
        pipe_final,
        param_grid=params,
        scoring='roc_auc',  # Используем ROC-AUC как метрику
        cv=5,               # 5-фолд кросс-валидация
        n_jobs=-1,          # Параллельные вычисления
        verbose=2
    )

    # Ищем лучшие гиперпараметры
    search.fit(X_train, y_train)

    # Сохраняем лучшую модель и её результат
    best_models[model_name] = search.best_estimator_
    best_scores[model_name] = search.best_score_

    print(f"Лучшая модель для {model_name}: {search.best_params_}")
    print(f"ROC-AUC для {model_name}: {search.best_score_:.4f}")
````

````python
# Удаляем признаки из обучающей и тестовой выборок
X_train_filtered = X_train.drop(columns=features_to_remove)
X_test_filtered = X_test.drop(columns=features_to_remove)

print(f"Форма данных до фильтрации: {X_train.shape}")
print(f"Форма данных после фильтрации: {X_train_filtered.shape}")
````

````python
# Пересоздаём пайплайны для каждого типа признаков
ohe_pipe_updated = Pipeline([
    ("SimpleImputer_ohe", SimpleImputer(strategy="most_frequent")),
    ("ohe", OneHotEncoder(drop="first", handle_unknown="ignore")),
])

# Обработчик ordinal признаков (если остались ordinal признаки)
if ord_columns_updated:
    ord_pipe_updated = Pipeline([
        ("SimpleImputer_ord", SimpleImputer(strategy="most_frequent")),
        ("ord", OrdinalEncoder(
            categories=[
                [
                    "Товары для детей",
                    "Домашний текстиль", 
                    "Косметика и аксесуары",
                    "Техника для красоты и здоровья",
                    "Кухонная посуда",
                    "Мелкая бытовая техника и электроника",
                ]
            ],
            handle_unknown="use_encoded_value",
            unknown_value=-1,
        )),
    ])

# Создаём обновлённый препроцессор
transformers_list = []

if ohe_columns_updated:
    transformers_list.append(("ohe", ohe_pipe_updated, ohe_columns_updated))

if ord_columns_updated:  
    transformers_list.append(("ord", ord_pipe_updated, ord_columns_updated))

if num_columns_updated:
    transformers_list.append(("num", StandardScaler(), num_columns_updated))

data_preprocessor_updated = ColumnTransformer(
    transformers_list,
    remainder="passthrough",
    verbose_feature_names_out=False,
)

print("Обновлённый препроцессор создан!")
````

````python
# Применяем препроцессор к обучающим данным для определения структуры
X_preprocessed_updated = data_preprocessor_updated.fit_transform(X_train_filtered)

print("Форма данных после обновлённого препроцессинга:", X_preprocessed_updated.shape)

# Формируем новую категориальную маску
categorical_mask_updated = []

# Добавляем маску для OHE признаков
if ohe_columns_updated:
    ohe_feature_names_updated = data_preprocessor_updated.named_transformers_["ohe"]["ohe"].get_feature_names_out()
    categorical_mask_updated.extend([True] * len(ohe_feature_names_updated))

# Добавляем маску для ординальных признаков  
if ord_columns_updated:
    categorical_mask_updated.extend([True] * len(ord_columns_updated))

# Добавляем маску для числовых признаков
if num_columns_updated:
    categorical_mask_updated.extend([False] * len(num_columns_updated))

# Проверяем длину маски
assert len(categorical_mask_updated) == X_preprocessed_updated.shape[1], "Длина categorical_mask не совпадает с количеством признаков"

print("Обновлённая категориальная маска:", categorical_mask_updated)
````

````python
# Создаём обновлённый финальный пайплайн
pipe_final_updated = ImbPipeline(
    steps=[
        ('preprocessor', data_preprocessor_updated),
        ('smote', SMOTENC(
            categorical_features=categorical_mask_updated,
            random_state=RANDOM_STATE
        )),
        ('models', LogisticRegression(
            max_iter=1000,
            C=0.01,
            penalty='l2', 
            solver='lbfgs',
            random_state=RANDOM_STATE
        ))
    ]
)

print("Обновлённый пайплайн создан успешно!")
````

### Обучение модели на новых данных

````python
# Обучаем модель на отфильтрованных данных
print("Обучение модели на отфильтрованных данных...")
pipe_final_updated.fit(X_train_filtered, y_train)

# Проверяем качество на кросс-валидации
roc_auc_scores = cross_val_score(
    pipe_final_updated,
    X_train_filtered,
    y_train,
    scoring='roc_auc',
    cv=5,
    n_jobs=-1
)

print(f"ROC-AUC на кросс-валидации (отфильтрованные данные): {roc_auc_scores.mean():.4f} (+/- {roc_auc_scores.std():.4f})")
````

ROC-AUC на кросс-валидации (отфильтрованные данные): 0.8991, ранее мы получили на тренировочных результатах 0,8993 - разница совсем незначительная, так что можно делать прогноз на тестовых и смотреть уже на метрику.
